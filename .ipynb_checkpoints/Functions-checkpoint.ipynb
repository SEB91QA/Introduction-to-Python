```python
def maximum(value1, value2, value3):
    '''Returns the highest value'''
    max_value = value1
    if value2 > max_value:
        max_value = value2
    if value3 > max_value:
        max_value = value3
    return max_value


maximum(12, 27, 36)
```

```python
maximum(12.3, 45.6, 9.7)
```

```python
maximum('yellow', 'red', 'orange')
```

```python
maximum(13.5, -3, 7)
```

```python
max('yellow', 'red', 'orange', 'blue', 'green')
```

```python
import random  # pseudorandom numbers

for roll in range(10):
    print(random.randrange(1, 7), end=' ')
```

```python
frequency1 = 0
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

for roll in range(6_000_000):
    face = random.randrange(1, 7)

    if face == 1:
        frequency1 += 1
    if face == 2:
        frequency2 += 1
    if face == 3:
        frequency3 += 1
    if face == 4:
        frequency4 += 1
    if face == 5:
        frequency5 += 1
    if face == 6:
        frequency6 += 1

print(f'Face{"Frequency":>13}')
print(f'{1:>4}{frequency1:>13}')
print(f'{2:>4}{frequency2:>13}')
print(f'{3:>4}{frequency3:>13}')
print(f'{4:>4}{frequency4:>13}')
print(f'{5:>4}{frequency5:>13}')
print(f'{6:>4}{frequency6:>13}')
```
