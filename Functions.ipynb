```python
def maximum(value1, value2, value3):
    '''Returns the highest value'''
    max_value = value1
    if value2 > max_value:
        max_value = value2
    if value3 > max_value:
        max_value = value3
    return max_value


maximum(12, 27, 36)
```

```python
maximum(12.3, 45.6, 9.7)
```

```python
maximum('yellow', 'red', 'orange')
```

```python
maximum(13.5, -3, 7)
```
