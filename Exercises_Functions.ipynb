```python
def mystery(x):
    '''Suma de los elementos pares'''
    y = 0

    for i in x:
        if i % 2 == 0:
            y += i
    return y


def calculate_product(args):
    p = 1
    for i in args:
        p *= i
    return p


l = [2, 4, 5, 7, 9, 2]
mystery(l)
```

```python
calculate_product(l)
```

```python
mystery([2, 4, 5, 7, 9, 1])
```

```python
def calc_slope(x1, y1, x2, y2):
    delta_x = x2 - x1
    delta_y = y2 - y1
    slope = delta_y/delta_x
    return slope


calc_slope(1, 1, 4, 5)
```

```python
calc_slope(*(1, 1), *(4, 5))
```
