```python
def mystery(x):
    '''Suma de los elementos pares'''
    y = 0

    for i in x:
        if i % 2 == 0:
            y += i
    return y


def calculate_product(args):
    p = 1
    for i in args:
        p *= i
    return p


l = [2, 4, 5, 7, 9, 2]
mystery(l)
```

```python
calculate_product(l)
```

```python
mystery([2, 4, 5, 7, 9, 1])
```

```python
def calc_slope(x1, y1, x2, y2):
    delta_x = x2 - x1
    delta_y = y2 - y1
    slope = delta_y/delta_x
    return slope


calc_slope(1, 1, 4, 5)
```

```python
calc_slope(*(1, 1), *(4, 5))
```

```python
def average(element, *args):
    return (sum(args) + element)/(1 + len(args))


average(4.00, *(4.00, 5.00))
```

```python
import random


def multiplication():
    num1 = random.randrange(1, 11)
    num2 = random.randrange(1, 11)

    print(f'The numbers are {num1} and {num2}')
    return num1, num2


def intelligence(num1, num2):

    return num1*(num2 + random.randrange(0, 2))
    # We need to add some balance. We can add or zero or 1


'''Product with intelligence'''

c = 'Yes'
v = 0
l = 0
e = 0
n = 0
while c == 'Yes':
    num1, num2 = multiplication()
    product = int(input('Please enter the product of numbers'))
    result_machine = intelligence(num1, num2)
    print(f"The machine's result is {result_machine}")

    if product == num1*num2 and result_machine != num1*num2:
        v += 1
        print('You win')

    elif product != num1*num2 and result_machine == num1*num2:
        l += 1
        print('You lose')

    elif product == num1*num2 and result_machine == num1*num2:
        e += 1
        print('Both win')

    else:
        print('Both answers are incorrect')

    n += 1
    c = input('Please enter Yes if want to keep playing')

print(f'You had {v} wins, {l} loses and {e} draws')
```
